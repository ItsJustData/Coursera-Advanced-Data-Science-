## Prepare the data sets for visualization

```python
def dropcol(df: pd.DataFrame) -> pd.DataFrame:
    '''drops the selected columns from each dataframe'''
    df.drop(['Province/State','Lat','Long'], axis=1, inplace=True)
    return df
```

```python
def stringcheck(df: pd.DataFrame) -> bool:
    '''checks that all column labels are of the string type'''
    boolean = all(isinstance(column, str) for column in df.columns)
    return boolean    
```

```python
print('\n',stringcheck(df_confirm_glb),'\n', stringcheck(df_death_glb),
      '\n', stringcheck(df_recover_glb))
```

```python
def setcountry(df: pd.DataFrame) -> pd.DataFrame:
    '''sets the country/region as the index'''
    df.set_index('Country/Region', inplace=True)
    return df
```

```python
setcountry(df_confirm_glb).head()
```

```python
def renametot(df: pd.DataFrame, name: str) -> pd.DataFrame:
    '''renames the final accumulative column to the total value'''
    df = df.rename(columns={'4/13/20':f'Total{name}'})
    return df
```

```python
df_confirm_glb = renametot(df_confirm_glb, 'Cases')
df_confirm_glb.head()
```

```python
df_death_glb = renametot(df_death_glb, 'Deaths')
df_death_glb.head()
```

```python
df_recover_glb = renametot(df_recover_glb, 'Recoveries')
df_recover_glb.head()
```

```python
def groupsum(df: pd.DataFrame) -> pd.DataFrame:
    '''groups by each country/region and adds their rows together'''
    df = df.groupby(['Country/Region']).sum()
    return df
```

```python
df_confirm_glb = groupsum(df_confirm_glb)
df_confirm_glb.head()
```

```python
df_death_glb = groupsum(df_death_glb)
df_death_glb.head()
```

```python
df_recover_glb = groupsum(df_recover_glb)
df_recover_glb.head()
```

## Sort the data by the top 10 totals of each dataframe

```python
def sort_ten(df: pd.DataFrame, name: str) -> pd.DataFrame:
    '''sorts the country/region by the top 10 in descending order'''
    df = df.sort_values(by=['Total{}'.format(name)], ascending=False)
    df = df.head(10)
    return df
```

```python
df_confirm_glb_t10 = sort_ten(df_confirm_glb, 'Cases')
df_confirm_glb_t10
```

```python
df_death_glb_t10 = sort_ten(df_death_glb, 'Deaths')
df_death_glb_t10
```

```python
df_recover_glb_t10 = sort_ten(df_recover_glb, 'Recoveries')
df_recover_glb_t10
```

```python
# transpose to set dates as the axis for visualization
df_confirm_glb_t10_tp = df_confirm_glb_t10.transpose()
df_confirm_glb_t10_tp.head()
```

```python
df_death_glb_t10_tp = df_death_glb_t10.transpose()
df_death_glb_t10_tp.head()
```

```python
df_recover_glb_t10_tp = df_recover_glb_t10.transpose()
df_recover_glb_t10_tp.head()
```

## Create a line plot for the top 10 countries of each dataframe

```python
def linplt(df: pd.DataFrame, name: str):
    ax = df.plot(kind='line', figsize=(10,6))
    ax.set_title('COVID-19 {} - Top 10 Countries (1-22-20 to 4-13-20)'.format(name), fontsize=16)
    ax.set_ylabel('Number of {}'.format(name), fontsize=14)
    ax.set_xlabel('Dates', fontsize=14)

    # sets the bottom y axis view limit to zero
    ax.set_ylim(bottom=0)

    labels=ax.get_xticklabels()

    # rotates labels to avoid text overlap
    plt.setp(labels, rotation = 45, horizontalalignment = 'right')

    # finds up to a max number of ticks at set locations
    ax.xaxis.set_major_locator(plt.MaxNLocator(10))
```

```python
linplt(df_confirm_glb_t10_tp, 'Confirmed Cases')
```

```python
linplt(df_death_glb_t10_tp, 'Confirmed Deaths')
```

```python
linplt(df_recover_glb_t10_tp, 'Confirmed Recoveries')
```

## Create a histogram for the top 10 countries of each dataframe and compare with the global data

```python
df_confirm_glb_t10['TotalCases'].head()
```

```python
count,bin_edges = np.histogram(df_confirm_glb_t10['TotalCases'])
# frequency count
print(count)
# bin ranges, default=10 bins (intervals of 55,003)
print(bin_edges)
```

```python
def histoplt(df1: pd.DataFrame, df2: pd.DataFrame, name: str):
    '''Create a histogram for the top 10 countries of each dataframe compared with the global data'''
    
    # plot both charts side by side for comparison
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(19,6))

    # set number of bins
    n_bins = 10

    ax1.hist(df1['Total{}'.format(name)], bins=n_bins, color='orange', alpha=0.5, ec='black')
    ax1.set_title('Total COVID-19 {} - Top 10 Countries (1-22-20 to 4-13-20)'.format(name), fontsize=16)
    ax1.set_ylabel('Number of Countries', fontsize=14)
    ax1.set_xlabel('Number of {}'.format(name), fontsize=14)

    ax2.hist(df2['Total{}'.format(name)], bins=n_bins, color='orange', alpha=0.5, ec='black')
    ax2.set_title('Total COVID-19 {} - Global (1-22-20 to 4-13-20)'.format(name), fontsize=16)
    ax2.set_ylabel('Number of Countries', fontsize=14)
    ax2.set_xlabel('Number of {}'.format(name), fontsize=14)

    matplotlib.rc('xtick', labelsize=12) 
    matplotlib.rc('ytick', labelsize=12) 
```

```python
histoplt(df_confirm_glb_t10, df_confirm_glb, 'Cases')
```

```python
count2,bin_edges2 = np.histogram(df_death_glb_t10['TotalDeaths'])
# frequency count
print(count2)
# bin ranges, default=10 bins (intervals of 2,069.6)
print(bin_edges2)
```

```python
histoplt(df_death_glb_t10, df_death_glb, 'Deaths')
```

```python
count3,bin_edges3 = np.histogram(df_recover_glb_t10['TotalRecoveries'])
# frequency count
print(count3)
# bin ranges, default=10 bins (intervals of 7,059.2)
print(bin_edges3)
```

```python
histoplt(df_recover_glb_t10, df_recover_glb, 'Recoveries')
```

## Create a scatter plot for the number of deaths

```python
df_death_glb_tp = df_death_glb.transpose()
df_death_glb_tp.reset_index(inplace=True)
df_death_glb_tp = df_death_glb_tp.rename(columns={'index':'Dates'})
df_death_glb_tp = df_death_glb_tp.replace('TotalDeaths', '4/13/20')
df_death_glb_tp.head()
```

```python
# create a dataframe for the sum of the total deaths for each date 
df_death_tot = pd.DataFrame(df_death_glb_tp.iloc[:,1:].sum(axis=1))
df_death_tot.columns=['TotalDeaths']                             
df_death_tot.head()
```

```python
fig, (ax) = plt.subplots(1, figsize=(10,6))

ax.scatter(x=df_death_glb_tp['Dates'], y=df_death_tot['TotalDeaths'], c='darkblue', alpha=0.5)
ax.set_title('Total COVID-19 Deaths - Global (1-22-20 to 4-13-20)', fontsize=16)
ax.set_ylabel('Number of Deaths', fontsize=14)
ax.set_xlabel('Dates', fontsize=14)

labels=ax.get_xticklabels()
plt.setp(labels, rotation = 45, horizontalalignment = 'right')
ax.xaxis.set_major_locator(plt.MaxNLocator(10))
```
